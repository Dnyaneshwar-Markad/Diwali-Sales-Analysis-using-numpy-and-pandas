```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.drop(['Status', 'unnamed1'], axis=1)
#df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python

df.head(10)
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.columns
```

```python
df=df.rename(columns={'User_ID':'Id','Cust_name':'Name','Marital_Status':'Married','Product_Category':'Product'})
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Analysis of data using graps and visual methods
```

```python
#for counting males and females in data
a = sns.countplot(x='Gender',data=df,hue='Gender')
for bar in a.containers:
    a.bar_label(bar)
```

```python
# Barchart for Gender vs Amount
b= df.groupby(['Gender'], as_index=False)['Amount'].sum()
sns.barplot(x='Gender', y='Amount', data=b, hue='Gender')
```

```python
#from above graph we can see the Females have a greater purchasing power than men
```

```python
# Age as sales factor
```

```python
a=sns.countplot(x='Age Group', data=df, hue='Age Group')
for bar in a.containers:
    a.bar_label(bar)
    
```

```python
a=sns.countplot(x='Age Group', data=df, hue='Gender')
for bar in a.containers:
    a.bar_label(bar)
```

```python
b= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=b, hue='Age Group')
```

```python
''' As we can see from the above graph the mejority of customers are of the 18-45 Age. The age group 26-35 is dominating the sales sector '''
```

```python
#Analysis of sales on the basis of State
```

```python
#State vs Sales/Amount
a= df.groupby(['State'], as_index=False)['Amount'].sum()
sns.barplot(x='State', y='Amount', data=a, hue='State')
sns.set({'figure.figsize':(15,5)})
```

```python
#Giving top 10 states in terms of Sales

a= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='State', y='Amount', data=a, hue='State')
sns.set({'figure.figsize':(15,5)})
```

```python
'From above we can see that most of the sales are from Uttar Pradesh, Maharashtra, Karnataka and Delhi respectively... '
```

```python
#Marital Status
```

```python
ax = sns.countplot(data = df, x = 'Married', hue='Married')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Married', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Married',y= 'Amount', hue='Gender')
```

```python
'From above graphs we can see that most of the buyers are married (women) and they have high purchasing power'
```

```python

# Occupation
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation', hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#top 10 occupation in sales
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue='Occupation')
```

```python
'From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector'
```

```python

# Product Category
```

```python
sns.set({'figure.figsize':(20,10)})
ax = sns.countplot(data = df, x = 'Product', hue='Product')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set({'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product',y= 'Amount', hue='Product')
```

```python
'From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category'
```

```python

# Specific Product Id's Sales
```

```python
#top 10 products
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(12,7)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue='Product_ID')
```

```python
'Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category'
```
